```python
import os
from langchain_huggingface import HuggingFaceEmbeddings
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Initialize the Embeddings
embeddings_1 = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

embeddings_2 = HuggingFaceEmbeddings(model_name="multi-qa-mpnet-base-dot-v1")

embeddings_3 = HuggingFaceEmbeddings(model_name='avsolatorio/GIST-Embedding-v0')

embeddings_4 = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5")

embeddings_5 = HuggingFaceEmbeddings(model_name="intfloat/e5-small")

embeddings_6 = HuggingFaceEmbeddings(model_name="dunzhang/stella_en_1.5B_v5")

text_pos_test = "The boy did not like the movie."
text_neg_test = "The boy did like the movie."

text_pos_test = "The outcome was favorable for the opposing team since they won."
text_neg_test = "The outcome was not favorable for the opposing team since they lost."
```

```python
def get_cos_sim(embedding_model, text_1, text_2):
    result_1 = embedding_model.embed_query(text_1)
    result_2 = embedding_model.embed_query(text_2)

    # Calculate cosine similarity
    similarity = cosine_similarity([result_1], [result_2])[0][0]
    return similarity

print(f"Cosine Similarity (all-MiniLM-L6-v2): {get_cos_sim(embeddings_1, text_pos_test, text_neg_test)}")
print(f"Cosine Similarity (multi-qa-mpnet-base-dot-v1): {get_cos_sim(embeddings_2, text_pos_test, text_neg_test)}")
print(f"Cosine Similarity (avsolatorio/GIST-Embedding-v0): {get_cos_sim(embeddings_3, text_pos_test, text_neg_test)}")
print(f"Cosine Similarity (BAAI/bge-small-en-v1.5): {get_cos_sim(embeddings_4, text_pos_test, text_neg_test)}")
print(f"Cosine Similarity (intfloat/e5-small): {get_cos_sim(embeddings_5, text_pos_test, text_neg_test)}")
print(f"Cosine Similarity (dunzhang/stella_en_1.5B_v5): {get_cos_sim(embeddings_6, text_pos_test, text_neg_test)}")
```
